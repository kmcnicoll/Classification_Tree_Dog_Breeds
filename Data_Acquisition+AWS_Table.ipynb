```python
DB_NAME = 'Dog_Breeds'

TABLES = {}
TABLES['Dogs'] = (
    "CREATE TABLE Dogs ("
    "  Name varchar(60) NOT NULL," #name of the breed
    "  Male_Min_Weight int(4) NOT NULL," #minimum expected weight for males
    "  Male_Max_Weight int(4) NOT NULL," #max expected weight for males
    "  Female_Min_Weight int(4) NOT NULL," #minimum expected wieght for females
    "  Female_Max_Weight int(4) NOT NULL," #maximum expected wieghtf or females
    "  Male_Height int(4) NOT NULL," #Average Height Males in withers inches
    "  Female_Height int(4) NOT NULL," #Average Height Females withers inches
    "  Exercise_Min_Min_Daily int(3) NOT NULL," #minimum exercise for breed in minutes daily
    "  Exercise_Max_Min_Daily int(3) NOT NULL," #upper time of expected exercise daily for breed IE 20-40 minutes daily. 40 is upper
    "  Energy_Level varchar(12) NOT NULL," #expected energy level of the breed
    "  Longevity_Min int(2) NOT NULL," #minimum expected life span for breed
    "  Longevity_Max int(2) NOT NULL," #upper expected life span for breed
    "  Tend_Drool varchar(8) ," # drooling tendancy for breed
    "  Tend_Snore varchar(8)," # snoring tendancy for breed
    "  Tend_Bark varchar(8) ," # barking tendancy for breed
    "  Tend_Dig varchar(8) ," # digging tendancy for breed
    "  Attention_Needs varchar(14)," # Attention needs for breed
    "  Grooming_Needs varchar(14) NOT NULL," # grooming care needed for coat
    "  AKC_Class varchar(40) NOT NULL," # American kennel club classification of breed
    "  UKC_Class varchar(40) NOT NULL," # United Kennel club classification for breed
    "  Prevalence varchar(22) NOT NULL," # Prevalance of breed AKA its popularity and commonness
    "  PRIMARY KEY (Name)"
    ") ENGINE=InnoDB")
```

```python
TABLES["Dogs"]
```

```python
from mysql.connector import errorcode
for table_name in TABLES:
    table_description = TABLES[table_name]
    try:
        print("Creating table {}: ".format(table_name), end='')
        cursor.execute(table_description)
    except mysql.connector.Error as err:
        if err.errno == errorcode.ER_TABLE_EXISTS_ERROR:
            print("already exists.")
        else:
            print(err.msg)
    else:
        print("OK")

cursor.close()
cnx.close()
```

```python
def db_insert(results): #Insert function to add rows to database
    ins= """INSERT INTO Dogs
    (Name, Male_Min_Weight, Female_Min_Weight, Male_Max_Weight, Female_Max_Weight,Male_Height, Female_Height, Exercise_Min_Min_Daily, Exercise_Max_Min_Daily,
    Energy_Level, Longevity_Min, Longevity_Max, Tend_Drool, Tend_Snore, Tend_Bark, Tend_Dig, Attention_Needs, Grooming_Needs, AKC_Class, UKC_Class, Prevalence) 
    VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)"""
    cursor.execute(ins, results)  
    cnx.commit()
    print(results)
```
